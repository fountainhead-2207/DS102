```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv(r'/home/check/DATA/university/yr3, hk1/DS102 - ML for Statistic/Đồ án/goodread/DS102/DATA/silver/silver_2.csv')  
```

```python
strat_cols = ['Genre', 'is_expert', 'Author_Rating']
n_iterations = 3
samples_per_iteration = 50
total_samples = n_iterations * samples_per_iteration
```

```python

sampled_df = df.groupby(strat_cols, group_keys=False).apply(
    lambda x: x.sample(frac=total_samples / len(df), random_state=42)
)

if len(sampled_df) < total_samples:
    extra = df.drop(sampled_df.index).sample(total_samples - len(sampled_df), random_state=42)
    sampled_df = pd.concat([sampled_df, extra])

batches = np.array_split(sampled_df.sample(frac=1, random_state=42), n_iterations)  # Shuffle rồi split
```

```python
for i, batch in enumerate(batches, start=1):
    os.makedirs(f'Guideline {i}',exist_ok=True)
    batch.to_csv(f'Guideline {i}/guideline_{i}.csv', index=False)
```

```python
all_batches = pd.concat(batches)
print("\nPhân bố tổng 150 samples:")
print(all_batches['Genre'].value_counts(normalize=True))
print(all_batches['is_expert'].value_counts(normalize=True))
print(all_batches['Author_Rating'].value_counts(normalize=True))
```
