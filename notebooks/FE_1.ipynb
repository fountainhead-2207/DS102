```python
genre_mapping = {
    'genre fiction': 'Fiction',
    'fiction': 'Fiction',
    'nonfiction': 'Nonfiction',
    'children': 'Children'
}

df['Genre_clean'] = df['Genre'].map(genre_mapping)

df['Genre'] = df['Genre_clean']
df.drop(columns=['Genre_clean'], inplace=True)
```

```python
book_level = df.groupby(['title', 'author']).agg(      
    worst_rank=('rank', 'max'),        
    mean_rank=('rank', 'mean'),        
    total_weeks=('total_weeks', 'max'),
    Units_Sold=('Units_Sold', 'sum'),
    Gross_Sales=('Gross_Sales', 'sum'),
    average_rating=('average_rating', 'mean'),
    ratings_count=('ratings_count', 'mean'),              
).reset_index()
```

```python
book_level.head()
```

```python
df_merged = df.merge(book_level, on=['title', 'author'], how='left', suffixes=('', '_agg'))
```

```python
df_merged.columns
```

```python
columns_to_replace = [
    'rank',  
    'week',
    'first_week',
    'date_added',         
    'total_weeks',      
    'Units_Sold',       
    'Gross_Sales',     
    'average_rating',    
    'ratings_count',   
]
df_merged.drop(columns=columns_to_replace, inplace=True, errors='ignore')
```

```python
agg_cols_with_suffix = [col for col in df_merged.columns if col.endswith('_agg')]
rename_dict = {col: col.replace('_agg', '') for col in agg_cols_with_suffix}
df_merged.rename(columns=rename_dict, inplace=True)
```

```python
df_merged['is_expert'] = df_merged.groupby('Genre')['n_votes'].transform(lambda x: (x >= x.quantile(0.90)).astype(int))
```

```python
df_merged.head()
```

```python
# number of unique titles
df_merged['title'].nunique()
```

```python
def balanced_representative_reviews(group):
    group = group.sort_values('n_votes', ascending=False)
    expert = group[group['is_expert'] == 1].head(2)
    public = group[group['is_expert'] == 0]
    if len(public) >= 3:
        public = public.sample(n=3, random_state=42)
    else:
        public = public  
    return pd.concat([expert, public]).head(5)
```

```python
df_reps_balanced = (
    df_merged.groupby(['title', 'author'], group_keys=False)
             .apply(balanced_representative_reviews)
)
```

```python
df_reps_balanced.shape
```

```python
strat_cols = ['Genre', 'Author_Rating', 'is_expert']

sample_df = df_reps_balanced.groupby(strat_cols, group_keys=False).apply(
    lambda x: x.sample(n=min(len(x), max(1, 2100 // len(df_reps_balanced) * len(x))), random_state=42)
)

print(f"Sau stratified sampling: {len(sample_df)} rows")

if len(sample_df) < 2100:
    remaining = df_reps_balanced.drop(sample_df.index) 
    needed = 2100 - len(sample_df)
    
    extra_n = min(needed, len(remaining))
    
    if extra_n > 0:
        extra = remaining.sample(n=extra_n, random_state=42)
        sample_df = pd.concat([sample_df, extra], ignore_index=False)
        print(f"Đã bổ sung thêm {extra_n} rows → Tổng: {len(sample_df)}")
    else:
        print("Không còn dữ liệu để bổ sung thêm!")
else:
    print("Đã đủ hoặc vượt 2100 rows từ stratified sampling.")

if len(sample_df) < 2100:
    final_extra = df_reps_balanced.sample(n=(2100 - len(sample_df)), random_state=43)
    sample_df = pd.concat([sample_df, final_extra])
    print(f"Bổ sung cuối cùng → Tổng: {len(sample_df)} rows")

sample_df = sample_df.reset_index(drop=True)

print("\nFinal sample shape:", sample_df.shape)
print("Phân bố is_expert:")
print(sample_df['is_expert'].value_counts(normalize=True))
print("\nPhân bố Genre:")
print(sample_df['Genre'].value_counts(normalize=True))
```

```python
final_column_order = [
    # 1. Metadata sách 
    'title',
    'author',
    'year',                   # Năm NYT bestseller
    'publication_year',       # Năm xuất bản
    'description',
    'Genre',
    
    # 2. Critical success (NYT rank + longevity + author reputation)
    'Author_Rating',          # Novice/Intermediate/Famous
    'total_weeks',            # Longevity trên NYT (rất quan trọng cho critical)
    'best_rank',              # Peak rank
    'worst_rank',             # Worst rank
    'mean_rank',              # Stability
    'debut_rank',             # Initial reception
    
    # 3. Popular success (từ Goodreads)
    'average_rating',         # Rating trung bình của sách
    'ratings_count',          # Số lượng rating (popular proxy)
    
    # 4. Review cá nhân + phân loại expert/public (cho labeling/NLP)
    'rating',                 # Rating của từng review
    'review_text',            # Nội dung review (core cho NLP)
    'n_votes',                # Số vote (dùng để tạo is_expert)
    'is_expert',              # Cột phân loại expert/public (vàng)
    
    # 5. Commercial success (target chính)
    'Units_Sold',             # Target chính
    'Gross_Sales',
    'Sale_Price',
    'Sales_Rank'
]

df = sample_df[final_column_order]
print(df.columns.tolist())
```
